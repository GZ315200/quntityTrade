```python
import pandas as pd
```

```python
ChinaBank = pd.read_csv('ChinaBank.csv', index_col='Date').iloc[:,1:]
```

```python
ChinaBank.head()
```

```python
ChinaBank.index=pd.to_datetime(ChinaBank.index)
```

```python
Close=ChinaBank.Close
```

```python
Close
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('date')
plt.ylabel('close')
plt.title('A line chart of China of Bank 2014 close price')
plt.grid(True, axis='y')
plt.plot(Close['2014'],c='r')
```

```python
Close.describe()
```

```python
a=[0,0,0,0]
for i in Close:
    if (i>2)&(i<=3):
        a[0]+=1
    elif (i>3)&(i<=4):
        a[1]+=1
    elif (i>4)&(i<=5):
        a[2]+=1
    else:
        a[3]+=1
```

```python
a
```

```python
plt.bar([2,3,4,5],height=a,width=0.5,bottom=1.0)
```

```python
plt.hist(Close, bins=12)
```
